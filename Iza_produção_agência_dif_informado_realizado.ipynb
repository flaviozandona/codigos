```python
# lê tabela
df = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='VIDA', nrows=11, header=0)
# filtra/seleciona colunas utilizáveis
df = df[['Junção', 'Real. Mês']]
# altera tipo da coluna
df['Junção'] = df['Junção'].astype(int)
# renomeia as colunas da tabela
df = df.rename(columns={'Real. Mês': 'Real. Mês VIDA'})
display(df)

df1 = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='RE', nrows=11, header=0)
df1 = df1[['Junção', 'Real. Mês']]
df1['Junção'] = df1['Junção'].astype(int)
df1 = df1.rename(columns={'Real. Mês': 'Real. Mês RE'})
display(df1)

df2 = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='RESIDENCIAL', nrows=11, header=0)
df2 = df2[['Junção', 'Real. Mês']]
df2['Junção'] = df2['Junção'].astype(int)
df2 = df2.rename(columns={'Real. Mês': 'Real. Mês RES'})
display(df2)

df3 = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='DENTAL PF', nrows=11, header=0)
df3 = df3[['Junção', 'Real. Mês']]
df3['Junção'] = df3['Junção'].astype(int)
df3 = df3.rename(columns={'Real. Mês': 'Real. Mês DENTAL PF'})
display(df3)

df4 = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='SAUDE DENTAL PJ', nrows=11, header=0)
df4 = df4[['Junção', 'Real. Mês']]
df4['Junção'] = df4['Junção'].astype(int)
df4 = df4.rename(columns={'Real. Mês': 'Real. Mês DENTAL PJ'})
display(df4)

df5 = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='AUTO', nrows=11, header=0)
df5 = df5[['Junção','Nome da Agência', 'Real. Mês']]
df5['Junção'] = df5['Junção'].astype(int)
df5 = df5.rename(columns={'Real. Mês': 'Real. Mês AUTO'})
display(df5)

# mostra informações da tabela
# df1.info()
```

```python
# junta os três dataframes criados
df1 = df1.merge(df, on='Junção')
df2 = df2.merge(df1, on='Junção')
df3 = df3.merge(df2, on='Junção')
df4 = df4.merge(df3, on='Junção')
df6 = df5.merge(df4, on='Junção')
display(df6)
```

```python
# lê tabela
df_informado = pd.read_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/relatorio_resumo_produtos_0104-2804.xlsx', sheet_name='INFORMADO', nrows=11, header=0)

# excluí a coluna
df_informado = df_informado.drop('Total geral', axis=1)

# substitui por zero valor nulo
df_informado = df_informado.fillna(0)

# renomeia coluna da tabela
df_informado = df_informado.rename(columns={'Agência': 'Nome da Agência', 'RE(Emp, Cond, Equip)': 'RE'})

# cria coluna 'Residencial' e soma as duas colunas ('Pré Formatado' 'Sob Medida') para unificar os produtos
# df_informado['Residencial'] = df_informado.apply(lambda x : x['Pré Formatado']+x['Sob Medida'], axis=1)
df_informado['Dental PF'] = df_informado['Dental PF Anual']+df_informado['Dental PF Mensal']
df_informado['Residencial'] = df_informado['Pré Formatado']+df_informado['Sob Medida']
df_informado['Vida'] = df_informado['Vida Mensal']+df_informado['Vida Único/Anual']
df_informado['Dental PJ'] = df_informado['Dental PJ']+df_informado['Saúde']

# cria coluna com posição no dataframe e insere valores dos indices
df_informado.insert(0, 'Junção', [1925,974,1589,6521,3152,2270,1436,2890,1973,2183,285])

# cria coluna com indices no final do dataframe
#df_informado['Junção'] = [1925,974,1589,6521,3152,2270,1436,2890,1973,2183,285]

# filtra as colunas desejadas
df_informado = df_informado[['Junção',	'Automóvel',	'Dental PJ',	'RE',	'Dental PF',	'Residencial','Vida']]

display(df_informado)

# display(df_x)
```

```python
df7 = df6.merge(df_informado, on='Junção')
# reindexa (reordena) as colunas do dataframe
df7 = df7.reindex(columns=['Junção',	'Nome da Agência',	'Real. Mês AUTO',	'Automóvel', 'Real. Mês DENTAL PJ', 'Dental PJ',	'Real. Mês DENTAL PF', 'Dental PF',	'Real. Mês RES',	'Residencial', 'Real. Mês RE', 'RE',	'Real. Mês VIDA', 'Vida'])

# cria as diferenças entre o Realizado - Informado
df7.insert(4, 'DIF AUTO', df7['Real. Mês AUTO']-df7['Automóvel'])
df7.insert(7, 'DIF DENTAL PJ', df7['Real. Mês DENTAL PJ']-df7['Dental PJ'])
df7.insert(10, 'DIF DENTAL PF', df7['Real. Mês DENTAL PF']-df7['Dental PF'])
df7.insert(13, 'DIF RESIDENCIAL', df7['Real. Mês RES']-df7['Residencial'])
df7.insert(16, 'DIF RE', df7['Real. Mês RE']-df7['RE'])
df7.insert(19, 'DIF VIDA', df7['Real. Mês VIDA']-df7['Vida'])
# df7 = df7.fillna(0)
display(df7)
df7.to_excel(r'/content/drive/My Drive/Colab Notebooks/Iza/agencia_informado_realizado.xlsx', index=False)
```
